# Logistic regression
In this notebook, we implement an algorithm of logistic regression with $n$ number of features.<br>
For the cost function, we use the cross entropy function,
$$C(\mathbf{x}) = -\frac{1}{m}\sum_{i}^{m} y^{(i)}\ln\sigma(\mathbf{a}\cdot\mathbf{x}^{(i)})+(1-y^{(i)})\ln(1-\sigma(\mathbf{a}\cdot\mathbf{x}^{(i)})),$$
where $\mathbf{x}^{(i)}\in\{x_0^{(i)},x_1^{(i)},\cdots,x_n^{(i)}\}$ and $y^{(i)}$ are the training data, $m$ the number of training data, and $\mathbf{a}\in\{a_0,a_1,\cdots,a_n\}$ the fitting parameters.

Here, $\sigma(z)$ denotes the sigmoid function,
$$\sigma(z) = \frac{1}{1+e^{-z}}$$

The fitting parameters are optimized using gradient descent,
$$
\mathbf{a}^{\mathrm{new}} = \mathbf{a} - \eta \nabla_{\mathbf{a}} C(\mathbf{x}) = \mathbf{a} - \frac{\eta}{m} \sum_i^{m}\left[ (\sigma(\mathbf{a}\cdot\mathbf{x}^{(i)})-y^{(i)}) \mathbf{x}^{(i)}\right].
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class Logistic_regression():
    """
    """
    def __init__(self, x, y):
        """
        Initializing the setup.
        x : (n+1)-dimensional input training data, x.shape = (n+1, m)
            the first row (x[0]) corresponds to the 0th order.
        y : 1d output training data, y.shape = m
        a : (n+1)-dimensional fitting parameters
        costs: cost as a function of iteration
        """
        self.N, self.M   = x.shape
        self.a      = np.random.randn(self.N)
        self.costs  = []
        
    def sigmoid(self,x):
        return 1/(1+np.exp(-(self.a).dot(x)))
    
    def cross_entropy(self,x,y):
        return np.average(np.nan_to_num(-y*np.log(self.sigmoid(x))-(1-y)*np.log(1-self.sigmoid(x))))
    
    def update(self,x,y,eta):
        self.a -= eta*np.average([(self.sigmoid(x)-y)*I for I in x],axis=1)
        
    def run(self,x,y,eta,N):
        """
        Iterating the update function as well as calculating the cost.
        """
        for i in range(N):
            self.update(x,y,eta)
            self.costs.append(self.cross_entropy(x,y))
def decision_boundary(a, x1):
    return -(a[0] + a[1]*x1)/(a[2])
```

```python
# getting data from a url
data = pd.read_csv('https://raw.githubusercontent.com/guoqi228/logistic_regression_matlab/master/ex2data1.txt',header=None)
X = np.array(data[[0,1]]).T

## mean-normalization
#for i in range(X.shape[0]):
#    X[i] = (X[i]-np.average(X[i]))/(np.max(X[i])-np.min(X[i]))

# simple rescaling
X /= 100
X = np.vstack([np.ones(X.shape[1]),X])
Y = np.array(data[2])
```

```python
LR = Logistic_regression(X,Y)
LR.run(X,Y,3,30000)
```

```python
# x1 = np.linspace(X[1].min(),X[1].max(),100)
x1 = np.linspace(0.2,1,100)

fig, (ax0,ax1) = plt.subplots(1,2,figsize = (10,4))
plt.subplots_adjust(wspace=0.5)
scatter = ax0.scatter(X[1],X[2],c=Y)
ax0.set_xlabel("Exam 1",fontsize=15)
ax0.set_ylabel("Exam 2",fontsize=15)
ax0.plot(x1,decision_boundary(LR.a,x1),label = 'decision boundary')
ax0.legend(handles=scatter.legend_elements()[0], labels=["Rejected","Admitted"])

ax1.loglog(LR.costs)
ax1.set_xlabel('# of iterations',fontsize=15)
ax1.set_ylabel('Cost',fontsize=15)
plt.show()
```
