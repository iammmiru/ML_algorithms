# K-mean algorithm

In this note, we implement K-mean algorithm, a popular unsupervised machine learning algorithm.

K-mean algorithm tries to minimize the following cost function,
$$
C = \frac{1}{m}\sum_i^m |\mathbf{x}^{(i)} - \mathbf{\mu}_{c^{(i)}}|^2,
$$
where $\mathbf{x}^{(i)}$ is the $i$th training data, and $\mathbf{\mu}_{c^{(i)}}$ is the centroid of cluster that the training data has been assigned to.

The algorithm minimizes the cost function via repeatedly iterating two steps:
1. $\mathbf{x}^{(i)}$ is assigned to a cluster that is closest to $\mathbf{\mu}_{k}$.
2. Each centroid $\mathbf{\mu}_{k}$ is updated to the average of positions that are assigned to the cluster $k$.<br>

The number of clusters is an input parameter $N$, and the initial centroid positions are set by randomly selecting $N$ training points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

class Kmean():
    
    def __init__(self, X, N):
        indices = (np.random.random(N)*X.shape[0]).astype(int)
        self.mu = X[indices]
        self.assignment = 0
        self.N = N
        
    def cost_function(self, X, Y):
        return 0.5*(np.linalg.norm(X-Y,axis = 1)**2)
    
    def assign(self, X):
        dist = []
        for m in self.mu:
            dist.append(self.cost_function(X, m))
        dist = np.array(dist).T
        self.assignment = np.argmin(dist, axis = 1)
    
    def mu_update(self, X):
        for i in range(self.N):
            index = np.where(self.assignment==i)
            self.mu[i] = np.average(X[index], axis =0)
    
    def fit(self,X,runs):
        for i in range(runs):
            self.assign(X)
            self.mu_update(X)
```

```python
n_samples = 4000
n_components = 3

X, y = make_blobs(
    n_samples=n_samples, centers=n_components, cluster_std=2
)
```

```python
model = Kmean(X,3)
model.fit(X,1000)
```

```python
fig, (ax0,ax1) = plt.subplots(1,2,figsize=(10,4))
ax0.scatter(X[:,0],X[:,1],c=model.assignment)
ax0.set_title('Clustering with K-mean')
ax0.scatter(model.mu[:,0],model.mu[:,1],s=50, c='red', marker = 'X', label = 'centeroids')
ax0.legend(fontsize = 13)
ax1.scatter(X[:,0],X[:,1],c=y)
ax1.set_title('Test data')
plt.show()
```
